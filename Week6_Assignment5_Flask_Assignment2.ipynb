Q1. Explain GET and POST methods.

GET and POST are two types of HTTP request methods used to interact with web applications. In Flask we need to use 'request' object to access data sent with GET and POST requests.
GET Method:-
1. GET method is used for retriving data from web server. When we make a GET request, we are asking the server to provide with some information, and this request is typically appended with URL as query parameter.
2. GET methods are idempotent, meaning they should not change server's state. They are safe to chached, bookmarked and shared.
3. In Flask, we can handle GET requests by using '@app.route' decorator. By default routes created with this decorator respond to GET requests.
POST Method:-
1. POST method is used for submitting data to the web server. It is typically used when we want to send data to the server such as form submissions, file uploads, or other data that should not be included in the URL.
2. POST requests can change server's state and are not considered idempotent, as they can have side effect on the server.
3. In Flask we can handle POST requests by specifying 'request' arguement when using '@app.route' decorator. For handeling POST request we need to include 'POST' in the list of methods.

Q2. Why is request used in Flask?

In FLask, the 'request' object is used to access and interact with data sent by cleint as a part of HTTP request. It is a bulit-in object provided by Flask and os crucial for handling incoming data, such as form data, coockies, query parameters and  request headers.
1. The primary purpose of 'request' object is to retrive data from incoming HTTP requests. This data can include form data, query parameters, coockies, request headers, file uploads.
2. The 'request' object allows us to validate and process incoming data. For example, we can check that the required form field are present and have expected values. This is essential for ensuring data intigrity and security in our application.
3. When implimenting user authentication, we can use 'request' object to check user credentials, user information and perform task like user login and logout.
4. The 'request' object allows us to create custom middleware function in Flask. This middleware function can inspect and modify request object before it reaches the view function, enabling tasks like request logging, authentication and authorization.

Q3. Why is redirect() used in Flask?

The 'redirect()' function is used to perform HTTP redirects. Redirects are useful for controling the flow of web application by redirecting user's browser to different URL either within the same application or to an external location. 
1. redirect() function can be used to guide the user to a different location if they have completed a specific task or if they have reached a particular endpoint.
2. redirect() is often used for user authentication and authorization. For example, after successful login we can redirect the user to their home page and if the user authentication fails we can redirect the user to some different URL.
3. If we want to display some one-time message to the user such as success message to the user after form submission, we can use 'flash' messages in conjuntion with 'redirect()' This message can be flashed and displayed on the redirected page.
4. redirect() function can be used to design nevigation buttons to navigate through the different parts of the website.

Q4. What are templates in Flask? Why is the render_template() function used?

In flask templates are the fundamental component of web Application to separate presentation(HTML) from application logic(Python code). Tamplates allow us to create dynamic and data driven web pages by embedding variables and control structures within the HTML. FLask uses template engine to render these templates and inject dynamic contents. The 'render_template()' function is used to render these templates.  
1. Templates are typically HTML files with placeholders for dynamic contents.
2. Flask supports various template engines, but the most common is 'Jinja2' template engine.
3. Templates can include variables enclosed in double curly braces, control structures like 'for' loop, 'while' loop, 'if' statements and template inheritance to create a consistent layout for our web page.
4. The 'render_template()' is a built-in function in Flask that is used to render HTML templates and return HTTP responses to client's browser.
5. 'render_template()' takes name of the template file as arguement, along with optional template variables.
6. The  'render_template()' function loads the specified template, processes it using template engine(Jinja2), and substitute any variable with their values
7. The rendered HTML is then sent to client's browser as a response.
